## 1ï¸âƒ£ map(): aplicar una funciÃ³n a todos los elementos

`map(funciÃ³n, iterable)` devuelve un iterador con los resultados de aplicar la funciÃ³n a cada elemento.

Ejemplo bÃ¡sico:

````python
numeros = [1, 2, 3, 4, 5]
cuadrados = list(map(lambda x: x**2, numeros))
print(cuadrados)
````

âœ… `map()` transforma cada elemento sin alterar la lista original.

## 3ï¸âƒ£ filter(): seleccionar elementos segÃºn una condiciÃ³n

`filter(funciÃ³n, iterable)` devuelve un iterador con los elementos que cumplen la condiciÃ³n.

Ejemplo:

````python
numeros = [5, 12, 18, 7, 3]
mayores_10 = list(filter(lambda x: x > 10, numeros))
print(mayores_10)
````

âœ… `filter()` mantiene solo los elementos que cumplen la expresiÃ³n booleana de la lambda.

## 5ï¸âƒ£ reduce(): combinar todos los elementos en un Ãºnico valor

`reduce(funciÃ³n, iterable)` aplica una funciÃ³n acumulativa para reducir una secuencia a un solo valor.
ðŸ‘‰ Hay que importarla desde `functools`.

Ejemplo:

````python
from functools import reduce

numeros = [1, 2, 3, 4]
producto = reduce(lambda x, y: x * y, numeros)
print(producto)
````

âœ… `reduce()` aplica la funciÃ³n acumulativamente: (((1Ã—2)Ã—3)Ã—4) â†’ 24.

## 6ï¸âƒ£ ðŸ§© Ejercicio 3 â€” Sumar precios totales

Dada la lista `precios = [12.5, 7.8, 19.9, 4.5]`, usa `reduce()` para calcular el total acumulado.

ðŸ’¡ *Pista:* comienza con `0` como valor inicial (`reduce(func, iterable, inicio)`).

````python
# Implementa aquÃ­ tu soluciÃ³n...
````

### âœ… SoluciÃ³n propuesta

````python
from functools import reduce
precios = [12.5, 7.8, 19.9, 4.5]
total = reduce(lambda acc, x: acc + x, precios, 0)
print('Total:', total)
````

## 7ï¸âƒ£ Combinando map(), filter() y reduce()

Estas tres funciones se pueden **encadenar** para crear flujos de datos complejos sin bucles.

Ejemplo: calcular la suma de los cuadrados de los nÃºmeros pares de una lista.

````python
numeros = [1, 2, 3, 4, 5, 6]
from functools import reduce

resultado = reduce(
    lambda acc, x: acc + x,
    map(lambda n: n**2, filter(lambda n: n % 2 == 0, numeros)),
    0
)
print('Suma de cuadrados pares:', resultado)
````

âœ… `filter()` selecciona â†’ `map()` transforma â†’ `reduce()` combina.

## 8ï¸âƒ£ ðŸ§© Ejercicio 4 â€” Flujo completo de datos

Crea una lista de productos con precios en euros:
```python
productos = [15, 23, 5, 80, 40]
```
1. Filtra los productos con precio > 10.
2. Aplica un 5% de descuento a los seleccionados.
3. Calcula el total de la compra.

ðŸ’¡ Usa `filter()`, `map()` y `reduce()` en un solo bloque.

````python
# Escribe aquÃ­ tu flujo funcional...
````

### âœ… SoluciÃ³n propuesta

````python
from functools import reduce
productos = [15, 23, 5, 80, 40]

total_compra = reduce(
    lambda acc, x: acc + x,
    map(lambda p: p * 0.95, filter(lambda p: p > 10, productos)),
    0
)

print('Total final con descuento:', total_compra)
````
