## 1ï¸âƒ£ Escenario

Tenemos una lista de **ventas diarias** en formato `(vendedor, importe)`.

```python
ventas = [
    ('Ana', 1200), ('Luis', 800), ('Carlos', 1500), ('Ana', 700), ('Luis', 950), ('Ana', 300)
]
```

Queremos obtener informaciÃ³n como:
- Total de ventas globales.
- Total por vendedor.
- Filtrar vendedores con media superior a 1000â‚¬.
- Verificar si **todos** los vendedores superaron un umbral mÃ­nimo semanal.

````python
ventas = [
    ('Ana', 1200), ('Luis', 800), ('Carlos', 1500), ('Ana', 700), ('Luis', 950), ('Ana', 300)
]
````

---
## 2ï¸âƒ£ ğŸ§© Ejercicio 1 â€” Total global de ventas

Usa `reduce()` para calcular el **total global** de ventas.

ğŸ’¡ *Pista:* suma los segundos elementos de cada tupla (`importe`).

````python
# Escribe aquÃ­ tu cÃ³digo para calcular el total global...
````

### âœ… SoluciÃ³n propuesta

````python
from functools import reduce
total = reduce(lambda acc, v: acc + v[1], ventas, 0)
print('ğŸ’° Total global de ventas:', total, 'â‚¬')
````

---
## 3ï¸âƒ£ ğŸ§© Ejercicio 2 â€” Total de ventas por vendedor

Genera un diccionario con el total acumulado de cada vendedor.

ğŸ’¡ *Pista:* puedes usar comprensiÃ³n de diccionario junto con `filter()` o un bucle `set()` sobre los nombres.

````python
# Escribe aquÃ­ tu implementaciÃ³n...
````

### âœ… SoluciÃ³n propuesta

````python
vendedores = {v[0] for v in ventas}
totales = {v: sum(x[1] for x in ventas if x[0] == v) for v in vendedores}
print('Totales por vendedor:', totales)
````

---
## 4ï¸âƒ£ ğŸ§© Ejercicio 3 â€” Filtrar vendedores destacados

Queremos los vendedores cuya **media de ventas** sea superior a 1000â‚¬.

ğŸ’¡ *Pista:* usa `map()` o comprensiÃ³n de diccionario con una condiciÃ³n.

````python
# Escribe aquÃ­ tu filtro de destacados...
````

### âœ… SoluciÃ³n propuesta

````python
promedios = {
    v: (sum(x[1] for x in ventas if x[0] == v) / len([x for x in ventas if x[0] == v]))
    for v in vendedores
}
destacados = {v: p for v, p in promedios.items() if p > 1000}
print('ğŸ… Vendedores destacados:', destacados)
````

---
## 5ï¸âƒ£ ğŸ§© Ejercicio 4 â€” Verificar rendimiento global

Usa `all()` para comprobar si **todos** los vendedores superan un umbral mÃ­nimo (por ejemplo, 500â‚¬ en media semanal).

ğŸ’¡ *Pista:* usa el diccionario de `promedios` anterior.

````python
# Escribe aquÃ­ tu verificaciÃ³n con all()...
````

### âœ… SoluciÃ³n propuesta

````python
rendimiento_ok = all(media > 500 for media in promedios.values())
print('Â¿Todos los vendedores cumplen?', rendimiento_ok)
````

---
## 6ï¸âƒ£ ï¿½ï¿½ Ejercicio 5 â€” Generar resumen con zip()

Crea dos listas: `nombres` y `medias` a partir del diccionario de promedios, y usa `zip()` para unirlas.

ğŸ’¡ *Pista:* convierte el zip a lista y ordÃ©nala por media descendente con `sorted()`.

````python
# Escribe aquÃ­ tu cÃ³digo con zip()...
````

### âœ… SoluciÃ³n propuesta

````python
nombres, medias = list(promedios.keys()), list(promedios.values())
resumen = sorted(zip(nombres, medias), key=lambda x: x[1], reverse=True)
print('ğŸ“Š Ranking de vendedores:')
for n, m in resumen:
    print(f'{n:10} â†’ {m:.2f}â‚¬')
````
