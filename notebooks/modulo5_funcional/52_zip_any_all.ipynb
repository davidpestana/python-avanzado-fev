## 1Ô∏è‚É£ zip(): combinar secuencias

`zip()` agrupa elementos de varias secuencias en tuplas paralelas.

Ejemplo b√°sico:

````python
nombres = ['Ana', 'Carlos', 'Bea']
edades = [25, 30, 28]

combinados = list(zip(nombres, edades))
print(combinados)
````

‚úÖ El resultado es una lista de tuplas con los pares correspondientes.

‚ö†Ô∏è Si las listas tienen longitudes distintas, `zip()` se detiene en la m√°s corta.

## 2Ô∏è‚É£ üß© Ejercicio 1 ‚Äî Fusionar dos listas en un diccionario

Dadas dos listas:

```python
claves = ['a', 'b', 'c']
valores = [10, 20, 30]
```
Usa `zip()` y comprensi√≥n de diccionario para combinarlas en:

```python
{'a': 10, 'b': 20, 'c': 30}
```

````python
# Escribe aqu√≠ tu c√≥digo...
````

### ‚úÖ Soluci√≥n propuesta

````python
claves = ['a', 'b', 'c']
valores = [10, 20, 30]
diccionario = {k: v for k, v in zip(claves, valores)}
print(diccionario)
````

## 3Ô∏è‚É£ any() y all(): comprobaciones l√≥gicas sobre iterables

- `any(iterable)` devuelve **True** si *al menos un elemento* es verdadero.
- `all(iterable)` devuelve **True** si *todos los elementos* son verdaderos.

Ejemplo:

````python
valores = [True, False, True]
print(any(valores))  # True
print(all(valores))  # False
````

‚úÖ Son equivalentes a expresiones l√≥gicas tipo `OR` y `AND`, pero aplicadas a colecciones completas.

## 4Ô∏è‚É£ üß© Ejercicio 2 ‚Äî Validar datos de usuario

Crea un diccionario `usuario = {'nombre': 'David', 'email': 'david@correo.com', 'edad': 0}`.

Valida que **todos los campos tengan valor** (no est√©n vac√≠os o cero) usando `all()`.

üí° *Pista:* usa `usuario.values()` para recorrer los valores.

````python
# Escribe tu validaci√≥n aqu√≠...
````

### ‚úÖ Soluci√≥n propuesta

````python
usuario = {'nombre': 'David', 'email': 'david@correo.com', 'edad': 0}
valido = all(usuario.values())
print('¬øUsuario v√°lido?', valido)
````

‚úÖ `all()` eval√∫a los valores como booleanos (`0`, `''` y `None` se consideran `False`).

## 5Ô∏è‚É£ Combinando zip(), any() y all()

Podemos combinar estas funciones para realizar validaciones o comparaciones de listas.

Ejemplo: verificar si dos listas son iguales elemento a elemento.

````python
a = [1, 2, 3, 4]
b = [1, 2, 5, 4]

iguales = all(x == y for x, y in zip(a, b))
print('¬øListas iguales?', iguales)
````

‚úÖ `zip()` empareja los elementos y `all()` confirma si todos los pares cumplen la condici√≥n.

## 6Ô∏è‚É£ üß© Ejercicio 3 ‚Äî Control de acceso

Crea una lista `roles_usuario = ['admin', 'ventas']` y otra `roles_requeridos = ['admin', 'compras']`.

Valida con `any()` si **al menos uno** de los roles del usuario est√° permitido.

üí° *Pista:* usa comprensi√≥n de listas junto con `any()`.

````python
# Escribe aqu√≠ tu validaci√≥n de roles...
````

### ‚úÖ Soluci√≥n propuesta

````python
roles_usuario = ['admin', 'ventas']
roles_requeridos = ['admin', 'compras']

acceso = any(r in roles_requeridos for r in roles_usuario)
print('¬øAcceso permitido?', acceso)
````
