En este notebook aprender√°s a **agrupar textos similares** sin usar etiquetas de clase, aplicando el algoritmo **KMeans** sobre vectores TF-IDF.

üìò **Objetivo:** descubrir temas o patrones comunes en un conjunto de textos.

Usaremos las siguientes etapas:
1. Preprocesado de texto
2. Vectorizaci√≥n con TF-IDF
3. Entrenamiento de un modelo KMeans
4. Interpretaci√≥n de los cl√∫steres

```python
# üöÄ Instalaci√≥n y librer√≠as
try:
    import nltk
    from sklearn.feature_extraction.text import TfidfVectorizer
    from sklearn.cluster import KMeans
except ImportError:
    import sys
    !{sys.executable} -m pip install nltk scikit-learn
    import nltk
    from sklearn.feature_extraction.text import TfidfVectorizer
    from sklearn.cluster import KMeans

nltk.download('stopwords', quiet=True)
from nltk.corpus import stopwords
stop_words_es = stopwords.words('spanish')

print('‚úÖ Librer√≠as cargadas correctamente.')
```

## 1Ô∏è‚É£ Dataset de ejemplo
Trabajaremos con frases relacionadas con tres posibles temas ocultos: *tecnolog√≠a, deportes y salud*.

```python
corpus = [
    'La inteligencia artificial revoluciona la medicina',
    'Los m√©dicos usan algoritmos para diagnosticar enfermedades',
    'El Real Madrid gana otro partido importante',
    'Los jugadores entrenan para mejorar su rendimiento',
    'Nuevos avances en vacunas y tratamientos',
    'El Bar√ßa ficha a una nueva estrella del f√∫tbol',
    'La tecnolog√≠a 5G cambiar√° la conectividad global',
    'Python y la ciencia de datos dominan el mercado laboral',
    'Los investigadores descubren una terapia innovadora',
    'El ciclismo mejora la salud cardiovascular'
]

print(f'üìö Total de textos: {len(corpus)}')
```

## 2Ô∏è‚É£ Vectorizaci√≥n TF-IDF
Convertimos los textos en vectores num√©ricos eliminando *stopwords* en espa√±ol.

```python
vectorizador = TfidfVectorizer(stop_words=stop_words_es)
X = vectorizador.fit_transform(corpus)
print(f'‚úÖ Matriz TF-IDF generada con {X.shape[0]} documentos y {X.shape[1]} t√©rminos.')
```

## 3Ô∏è‚É£ Entrenamiento de KMeans
Entrenaremos un modelo de **3 cl√∫steres** (uno por cada tema esperado).

```python
modelo = KMeans(n_clusters=3, random_state=42, n_init=10)
modelo.fit(X)

print('‚úÖ Modelo KMeans entrenado correctamente.')
```

## 4Ô∏è‚É£ Exploraci√≥n de resultados
Veamos los textos agrupados seg√∫n el cl√∫ster asignado.

```python
etiquetas = modelo.labels_

for i in range(3):
    print(f'\nüîπ **Cluster {i}**')
    for texto in [corpus[j] for j in range(len(corpus)) if etiquetas[j] == i]:
        print('   -', texto)
```
