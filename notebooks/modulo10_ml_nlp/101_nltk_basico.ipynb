# üß† 10.1 ‚Äî Procesamiento del Lenguaje Natural con NLTK

En este notebook aprender√°s a usar **NLTK (Natural Language Toolkit)**, una de las librer√≠as m√°s cl√°sicas para **procesar texto en lenguaje natural**.

Exploraremos conceptos clave como:
- Tokenizaci√≥n de texto.
- Eliminaci√≥n de stopwords.
- Lematizaci√≥n y stemming.
- Frecuencia de palabras.

üìò *Objetivo:* limpiar, normalizar y analizar textos de forma b√°sica antes de aplicarlos en modelos de Machine Learning.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from collections import Counter

# Descargar recursos necesarios solo la primera vez
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')

print('‚úÖ NLTK cargado correctamente.')
```

---
## 1Ô∏è‚É£ Tokenizaci√≥n de texto

Dividimos un texto en **palabras individuales** (tokens) para poder analizarlas de forma independiente.

```python
texto = "El aprendizaje autom√°tico es fascinante. Permite que las m√°quinas aprendan de los datos."

tokens = word_tokenize(texto.lower())
print(tokens)
```

---
## 2Ô∏è‚É£ Eliminaci√≥n de Stopwords

Las *stopwords* son palabras muy comunes (como 'el', 'de', 'que') que normalmente no aportan significado relevante al an√°lisis.

```python
stop_words = set(stopwords.words('spanish'))

tokens_filtrados = [w for w in tokens if w.isalpha() and w not in stop_words]
print(tokens_filtrados)
```

---
## 3Ô∏è‚É£ Lematizaci√≥n y Stemming

Ambos procesos **reducen las palabras a su forma base**:
- *Stemming:* recorta palabras a su ra√≠z (m√°s r√°pido pero menos preciso).
- *Lematizaci√≥n:* devuelve la palabra base usando un diccionario ling√º√≠stico.

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

for palabra in ['running', 'studies', 'machines', 'learning']:
    print(f"{palabra:10s} ‚Üí stem: {stemmer.stem(palabra):10s} | lemma: {lemmatizer.lemmatize(palabra)}")
```

---
## 4Ô∏è‚É£ üß© Ejercicio ‚Äî Limpieza completa de texto

Implementa una funci√≥n que:
1. Reciba un texto.
2. Lo tokenice y elimine stopwords.
3. Devuelva la lista de palabras lematizadas.

üëâ *Tip:* combina los pasos anteriores en una funci√≥n.

```python
# ‚úèÔ∏è Tu c√≥digo aqu√≠...
```

### ‚úÖ Soluci√≥n propuesta

```python
def limpiar_texto(texto):
    tokens = word_tokenize(texto.lower())
    stop_words = set(stopwords.words('spanish'))
    lemmatizer = WordNetLemmatizer()
    filtradas = [lemmatizer.lemmatize(t) for t in tokens if t.isalpha() and t not in stop_words]
    return filtradas

texto_ejemplo = "Las m√°quinas est√°n aprendiendo r√°pidamente gracias a los datos disponibles."
print(limpiar_texto(texto_ejemplo))
```

---
## 5Ô∏è‚É£ Frecuencia de palabras

Podemos contar las palabras m√°s frecuentes con `collections.Counter`.

```python
tokens_limpios = limpiar_texto(texto)
frecuencias = Counter(tokens_limpios)
print(frecuencias.most_common(5))
```

---
## 6Ô∏è‚É£ üß© Ejercicio ‚Äî Visualizar las palabras m√°s frecuentes

Usa `matplotlib` para crear un gr√°fico de barras con las 10 palabras m√°s comunes.

üëâ *Pista:* usa `frecuencias.most_common(10)` y construye dos listas (`palabras`, `conteos`).

```python
# ‚úèÔ∏è Tu c√≥digo aqu√≠...
```

### ‚úÖ Soluci√≥n propuesta

```python
import matplotlib.pyplot as plt

palabras, conteos = zip(*frecuencias.most_common(10))
plt.bar(palabras, conteos)
plt.title('Palabras m√°s frecuentes')
plt.xticks(rotation=45)
plt.show()
```
