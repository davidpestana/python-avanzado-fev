# üìä 8.1 ‚Äî Gr√°ficos con Matplotlib y Seaborn

En este notebook aprender√°s los fundamentos de la **visualizaci√≥n de datos** en Python utilizando:

- üìà **Matplotlib**: la librer√≠a base para crear gr√°ficos en 2D.
- üé® **Seaborn**: una capa m√°s avanzada que facilita gr√°ficos estad√≠sticos con estilos atractivos.

üí° *Objetivo:* comprender las herramientas esenciales para representar datos de forma clara y est√©tica.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Estilos globales de Seaborn
sns.set_theme(style='whitegrid')
print('‚úÖ Notebook 8.1 ‚Äî Matplotlib y Seaborn cargado correctamente.')
```

---
## 1Ô∏è‚É£ Cargar datos de ejemplo

Usaremos el dataset `ventas.csv` (modificado previamente) para visualizar patrones en precios, unidades y beneficios.

```python
df = pd.read_csv('../../datasets/ventas.csv')
df.rename(columns={'cantidad': 'unidades', 'precio': 'precio_unitario'}, inplace=True)
df['ingresos'] = df['unidades'] * df['precio_unitario']
df.head()
```

---
## 2Ô∏è‚É£ Primer gr√°fico con Matplotlib

Matplotlib permite crear gr√°ficos personalizados con bajo nivel de control.

Vamos a representar el n√∫mero de unidades vendidas por producto.

```python
ventas_por_producto = df.groupby('producto')['unidades'].sum()

plt.figure(figsize=(6,4))
plt.bar(ventas_por_producto.index, ventas_por_producto.values, color='teal')
plt.title('Unidades vendidas por producto')
plt.xlabel('Producto')
plt.ylabel('Unidades')
plt.show()
```

‚úÖ Observa c√≥mo usamos `plt.figure()`, `plt.bar()` y etiquetas para construir el gr√°fico paso a paso.

---
## 3Ô∏è‚É£ üß© Ejercicio ‚Äî Gr√°fico de l√≠neas temporal

Usa `plt.plot()` para visualizar la evoluci√≥n de ingresos totales por d√≠a (`fecha`).

üëâ *Pista:* agrupa por fecha y suma la columna `ingresos`.

```python
# ‚úèÔ∏è Escribe tu c√≥digo aqu√≠...
```

### ‚úÖ Soluci√≥n propuesta

```python
df['fecha'] = pd.to_datetime(df['fecha'])
ingresos_dia = df.groupby('fecha')['ingresos'].sum()

plt.figure(figsize=(7,4))
plt.plot(ingresos_dia.index, ingresos_dia.values, marker='o', color='royalblue')
plt.title('Evoluci√≥n de ingresos diarios')
plt.xlabel('Fecha')
plt.ylabel('Ingresos (‚Ç¨)')
plt.show()
```

---
## 4Ô∏è‚É£ Visualizaci√≥n con Seaborn

Seaborn simplifica los gr√°ficos con funciones de m√°s alto nivel y estilos predefinidos.

Ejemplo: gr√°fico de dispersi√≥n (`scatterplot`) entre precio y unidades.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='precio_unitario', y='unidades', hue='producto', s=100)
plt.title('Relaci√≥n entre precio y unidades vendidas')
plt.show()
```

‚úÖ Seaborn maneja autom√°ticamente los colores, la leyenda y el estilo del gr√°fico.

---
## 5Ô∏è‚É£ üß© Ejercicio ‚Äî Comparar ingresos por producto (gr√°fico de barras Seaborn)

üëâ *Pista:* usa `sns.barplot(x='producto', y='ingresos', data=df)`.

```python
# üß© Tu c√≥digo aqu√≠...
```

### ‚úÖ Soluci√≥n propuesta

```python
plt.figure(figsize=(6,4))
sns.barplot(
    data=df,
    x='producto',
    y='ingresos',
    estimator=np.sum,
    errorbar=None,       
    hue='producto',      # Necesario para usar palette
    palette='viridis',
    legend=False         # Evita que aparezca la leyenda repetida
)
plt.title('Ingresos totales por producto')
plt.xlabel('Producto')
plt.ylabel('Ingresos (‚Ç¨)')
plt.show()
```

---
## 6Ô∏è‚É£ Personalizaci√≥n de estilos

Podemos cambiar temas, colores o a√±adir anotaciones:

```python
sns.set_style('darkgrid')
plt.figure(figsize=(6,4))
sns.lineplot(data=df, x='fecha', y='ingresos', hue='producto', marker='o')
plt.title('Ingresos por producto en el tiempo')
plt.show()
```
